```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import pandas as pd
import numpy as np
import matplotlib.cm as cm
```

```python
X = np.loadtxt('./gcn_embeddings.txt')
```

```python
X_embedded = TSNE(n_components=2).fit_transform(X)
```

```python
colors = cm.rainbow(np.linspace(0, 1, 10))
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=7, random_state=99).fit(X_embedded)
```

```python
import matplotlib.pyplot as plt
for i in range(X_embedded.shape[0]):
    plt.scatter(X_embedded[i][0],X_embedded[i][1],color=colors[int(kmeans.labels_[i])],s=2)
plt.show()
```

```python
node_data = pd.read_table('./data/cora/cora.content', header=None)
ids=node_data[1434].astype("category")
```

```python
a=0
b=0
for i in range(X_embedded.shape[0]):
    for j in range(X_embedded.shape[0]):
        if ids[i]==ids[j] and kmeans.labels_[i]==kmeans.labels_[j]:
            a=a+1
        if ids[i]!=ids[j] and kmeans.labels_[i]!=kmeans.labels_[j]:
            a=a+1
        b=b+1
    print(i)
```
